```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
product = pd.read_csv('result.csv')

# product.head()
plt.plot(product['time'].values, product['A'].values,'r-', label='A')
plt.plot(product['time'].values, product['B'].values,'g-', label='B')
plt.plot(product['time'].values, product['P12'].values,'m--', label='P12')
plt.plot(product['time'].values, product['P23'].values,'b--', label='P23')
plt.title('total product')
plt.legend()
plt.show()
```

```python
average_periods = 60
rate = product.diff(periods=average_periods)
fig, axs = plt.subplots(1,2, figsize=(10,5), constrained_layout=True)
for ax in axs.flatten():
    ax.plot(product['time'].values, rate['A'].values,'r-', label='A')
    ax.plot(product['time'].values, rate['B'].values,'g-', label='B')
    # plt.plot(product['time'].values, rate['P12'].values,'m-', label='P12')
    # plt.plot(product['time'].values, rate['P23'].values,'b-', label='P23')
    ax.set_title('rate')
    # plt.ylim([0,])
    ax.legend()
```
